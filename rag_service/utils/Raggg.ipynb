```python
import os
from PyPDF2 import PdfReader
```

```python
def extract_text(pdf_path):
    reader = PdfReader(pdf_path)

    text = ""

    for page in reader.pages:
        page_text = page.extract_text()

        if page_text:
            text += page_text + "\n"

    return text
```

```python
pdf_files = [
    {
        "path": "Grade 10 ICT - Lesson 01 - Module 1.pdf",
        "module": "1.1",
        "title": "Introduction to ICT"
    },

    {
        "path": "Grade 10 ICT - Lesson 01 - Module 2.pdf",
        "module": "1.2",
        "title": "Applications of ICT in Daily Life"
    },

    {
        "path": "Grade 10 ICT - Lesson 01 - Module 3.pdf",
        "module": "1.3",
        "title": "Benefits and Challenges of ICT"
    }
]
```

```python
documents = []

for pdf in pdf_files:

    text = extract_text(pdf["path"])

    documents.append({
        "text": text,
        "module": pdf["module"],
        "title": pdf["title"],
        "source": pdf["path"]
    })

print("Documents Loaded :", len(documents))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(

    chunk_size=800,

    chunk_overlap=150,

    separators=["\n\n","\n",". "," ",""]

)
```

```python
texts = []

metadatas = []

for doc in documents:

    chunks = splitter.split_text(doc["text"])

    for chunk in chunks:

        texts.append(chunk)

        metadatas.append({

            "module": doc["module"],

            "title": doc["title"],

            "source": doc["source"]

        })

print("Total Chunks :", len(texts))
```

```python
embedding_model = HuggingFaceEmbeddings(

    model_name="sentence-transformers/all-MiniLM-L6-v2"

)
```

```python
embeddings = embedding_model.embed_documents(texts)
```

```python
def retrieve_context(question, module=None, k=5):

    if module:

        docs = vectordb.similarity_search(

            question,

            k=k,

            filter={"module":module}

        )

    else:

        docs = vectordb.similarity_search(

            question,

            k=k

        )

    context = "\n\n".join([doc.page_content for doc in docs])

    return context
```

```python
def build_rag_prompt(question, score, module):

    context = retrieve_context(question, module)

    prompt = f"""
You are an expert Grade 10 ICT teacher.

Student Performance

Module : {module}

Score : {score}%

Learning Material

{context}

Instructions

Generate a personalized study note.

If score is above 75%

- Give only a short revision note.
- Mention important facts only.

If score is between 50% and 75%

- Give a medium-length explanation.
- Explain key concepts.
- Include simple examples.

If score is below 50%

- Give a detailed explanation.
- Explain every important concept.
- Highlight common mistakes.
- Give examination tips.

Use only the learning material provided.

Do not invent new information.

Question

{question}
"""

    return prompt
```

```python
def retrieve_module_context(module):

    docs = vectordb.similarity_search(
        query="important concepts",
        filter={"module": module},
        k=8
    )

    return "\n\n".join([doc.page_content for doc in docs])
```

```python
def build_prompt(module, score):

    context = retrieve_module_context(module)

    prompt = f"""
You are an experienced Grade 10 ICT teacher.

A student completed a quiz.

Module : {module}

Quiz Score : {score}%

Below is the official learning material.

========================

{context}

========================

Your task is to create a PERSONALIZED STUDY PLAN.

The study plan MUST contain exactly these sections.

--------------------------------------------------

1. Performance Analysis

Explain the student's performance.

Mention whether this module is

• Strong

• Average

• Weak

--------------------------------------------------

2. Personalized Study Note

Generate a study note ONLY using the retrieved content.

Do NOT add external information.

If score >75%

Generate a concise revision note.

If score between 50 and 75

Generate a medium explanation.

If score below 50

Generate a detailed explanation.

--------------------------------------------------

3. Topics to Focus

List the most important topics.

--------------------------------------------------

4. Study Time Recommendation

Recommend study duration.

Example

30 Minutes

1 Hour

2 Hours

3 Hours

depending on performance.

--------------------------------------------------

5. Practice Questions

Generate practice questions according to score.

Score above 75

5 questions

Score 50-75

10 questions

Score below 50

15 questions

Questions should include

• Multiple Choice

• Short Answer

• Structured Questions

--------------------------------------------------

6. Motivation

End with a short motivational message.

Only use the retrieved learning material.

"""
    return prompt
```
